```python
import math
import matplotlib.pyplot as plt
import numpy as np
# not complete and will need a revision 
# Function to calculate FOV for a sensor
def calculate_fov(sensor_width, focal_length):
    fov = 2 * math.atan(sensor_width / (2 * focal_length)) * (180 / math.pi)
    return fov

# Function to calculate GSD for a sensor at a given altitude and FOV
def calculate_gsd(altitude, fov_degrees):
    fov_radians = math.radians(fov_degrees)
    gsd = altitude * math.tan(fov_radians / 2)
    return gsd

#Inputs for sensor specs
sensor_width = 4.3  # in mm (only the active area)
focal_length = 11.5  # in mm

#Calculate FOV
fov_degrees = calculate_fov(sensor_width, focal_length)
print(f"FOV: {fov_degrees:.2f} degrees")

#GSD calculation
altitude = 500  # in meters
fov_degrees = 20  # in degrees

#Calculate GSD
gsd = calculate_gsd(altitude, fov_degrees)
print(f"GSD: {gsd:.2f} meters per pixel")

#Plot FOV vs Altitude
altitudes = np.linspace(100, 2000, 50)  # Create a range of altitudes
gsd_values = [calculate_gsd(alt, fov_degrees) for alt in altitudes]

plt.figure(figsize=(10, 6))
plt.plot(altitudes, gsd_values, label=f"FOV = {fov_degrees} degrees")
plt.title('Ground Sampling Distance (GSD) vs Altitude')
plt.xlabel('Altitude (m)')
plt.ylabel('GSD (m/pixel)')
plt.grid(True)
plt.legend()
plt.show()

# Plot FOV for different focal lengths
focal_lengths = np.linspace(5, 50, 50)
fov_values = [calculate_fov(sensor_width, fl) for fl in focal_lengths]

plt.figure(figsize=(10, 6))
plt.plot(focal_lengths, fov_values, label=f"Sensor width = {sensor_width} mm")
plt.title('Field of View (FOV) vs Focal Length')
plt.xlabel('Focal Length (mm)')
plt.ylabel('FOV (degrees)')
plt.grid(True)
plt.legend()
plt.show()
```
